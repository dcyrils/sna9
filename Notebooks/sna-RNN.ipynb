```
def generator(data, min_index, max_index, lookback, delay,  shuffle, batch_size, step):
  if max_index is None:
    max_index = len(data) - delay - 1
    
  i = min_index + lookback
  
  while 1:
    if shuffle:
      rows = np.random.randint(min_index + lookback, max_index, size=batch_size)
      
    else:
      if i + batch_size >= max_index:
        i = min_index + lookback
        
      rows = np.arange(i, min(i + batch_size, max_index))
      
      i += len(rows)
      
    samples = np.zeros((len(rows), lookback // step, data.shape[-1]))
    
    targets = np.zeros((len(rows),))
    for j, row in enumerate(rows):
      indices = range(row - lookback, row, step)
      samples[j] = data[indices]
      targets[j] = data[row + delay][25]   # y
    yield samples, targets
```

```
lookback = 400000       # среднее количество измерений за день
delay = 400000     # цель - один день 
step = 20000            # шаг - час 
batch_size = 1000
TG_max = 4000000
VG_max = 5900000
```

```
# generators
train_gen = generator(data, min_index=0, max_index=TG_max, lookback=lookback, delay=delay,
            shuffle=True, step=step, batch_size=batch_size)
val_gen = generator(data, min_index=TG_max+1, max_index=VG_max, lookback=lookback, delay=delay,
            shuffle=False, step=step, batch_size=batch_size)
test_gen = generator(data, min_index=VG_max+1, max_index=None, lookback=lookback, delay=delay,
            shuffle=False, step=step, batch_size=batch_size)

val_steps = (VG_max - (TG_max+1) - lookback) // batch_size
test_steps = (len(data) - (VG_max+1) - lookback) // batch_size
```

```
from keras.models import Sequential
from keras import layers
from keras.optimizers import RMSprop
```

```
model = Sequential()
model.add(layers.Embedding(8000000, 32, input_length=data.shape[1]))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.MaxPooling1D(pool_size=5, strides=2))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
model.add(layers.Dense(1))
#model.summary()
model.compile(optimizer=RMSprop(lr=1e-4), loss='binary_crossentropy', metrics=['acc'])
history = model.fit(data, y,
                    epochs=5,
                    batch_size=4096,
                    validation_split=0.2)
```
